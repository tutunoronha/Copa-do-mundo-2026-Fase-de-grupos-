```python
import pandas as pd
import matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/tutunoronha/Copa-do-mundo-2026-Fase-de-grupos-/main/copa2026_fase_grupos.csv"
df = pd.read_csv(url)
```

```python
df.columns
```

```python
df.shape
```

```python
df["Aproveitamento_%"] = (df["Pts"] / (df["PJ"] * 3) * 100).round(1)
df[["Selecao", "Grupo", "Pts", "Aproveitamento_%"]].head(10)
```

```python
ranking = df.sort_values(
    by=["Pts", "SG", "GM"], ascending=False
).reset_index(drop=True)
ranking.index = ranking.index + 1   # ranking começando em 1
ranking[["Selecao", "Grupo", "Pts", "VIT", "E", "DER", "SG", "Aproveitamento_%"]].head(15)
```

```python
classificados = df[df["Posicao"] <= 2].sort_values(["Grupo", "Posicao"])
classificados[["Grupo", "Posicao", "Selecao", "Pts", "SG"]]
```

```python
print("=== TOP 5 ATAQUES (gols marcados) ===")
print(df.sort_values("GM", ascending=False)[["Selecao", "GM"]].head(5).to_string(index=False))
print("\n=== TOP 5 DEFESAS (gols sofridos) ===")
print(df.sort_values("GC")[["Selecao", "GC"]].head(5).to_string(index=False))
```

```python
resumo_grupos = df.groupby("Grupo").agg(
    Gols_Marcados=("GM", "sum"),
    Media_Gols=("GM", "mean"),
    Pts_Total=("Pts", "sum"),
).round(2).sort_values("Gols_Marcados", ascending=False)
resumo_grupos
```

```python
top10 = ranking.head(10)
plt.figure(figsize=(10, 5))
plt.barh(top10["Selecao"], top10["Pts"], color="seagreen")
plt.gca().invert_yaxis()
plt.xlabel("Pontos")
plt.title("Top 10 seleções por pontos — Fase de Grupos Copa 2026")
plt.tight_layout()
plt.show()
```

```python
df.groupby("Grupo")["Pts"].sum().plot(kind="line")
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df["GM"], df["GC"], color="steelblue")
for _, r in df.iterrows():
    plt.annotate(r["Selecao"], (r["GM"], r["GC"]), fontsize=7, alpha=0.7)
plt.xlabel("Gols marcados")
plt.ylabel("Gols sofridos")
plt.title("Ataque x Defesa")
plt.tight_layout()
plt.show()
```

```python
ordenado = df.sort_values("SG", ascending=False)
cores = ["seagreen" if x >= 0 else "indianred" for x in ordenado["SG"]]
plt.figure(figsize=(11, 9))
plt.barh(ordenado["Selecao"], ordenado["SG"], color=cores)
plt.gca().invert_yaxis()
plt.xlabel("Saldo de gols")
plt.title("Saldo de gols por seleção")
plt.tight_layout()
plt.show()
```
